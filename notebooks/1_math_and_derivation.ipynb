# Двухслойный перцептрон: архитектура и вывод градиентов

Рассмотрим нейронную сеть с:
- Входным слоем размерности $d = 2$,
- Одним скрытым слоем с $h = 10$ нейронами и активацией **ReLU**,
- Выходным слоем с $C = 2$ классами и функцией **softmах**.

Мы выведем градиенты для обучения с помощью **градиентного спуска** и **категориальной кросс-энтропии**.
